# 분류 모델 성능 평가

`sklearn.metrics` 패키지
- `confusion_matrix`: 혼동 행렬 생성<br>
- `classification_report`: 정밀도, 재현율, F1 점수
- `roc_curve`: FP Rate와 TP Rate (재현율) 계산 자동화
- `auc`: ROC curve의 면적 계산

```python
from sklearn.metrics import confusion_matrix

y_true = [1, 0, 1, 1, 1, 0, 1, 0, 0, 0, 1, 1, 1, 0, 1]
y_pred = [1, 0, 1, 1, 1, 0, 1, 1, 0 ,0 ,0, 1, 1, 0, 1]
confusion_matrix(y_true, y_pred)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_true, y_pred, target_names=['class 0', 'class 1']))
```

```python
from sklearn.metrics import roc_curve

# pos_label은 기준값
fpr, tpr, thresholds = roc_curve(y_true, y_pred, pos_label=1)
fpr, tpr, thresholds
```

```python
import matplotlib.pyplot as plt

plt.plot(fpr, tpr)
plt.xlabel('FP Rate')
plt.ylabel('TP Rate(Recall)')
plt.show()
```

```python
from sklearn.metrics import auc
round(auc(fpr, tpr), 4)
```

```python
from sklearn.metrics import roc_auc_score
round(roc_auc_score(y_true, y_pred), 4)
```
