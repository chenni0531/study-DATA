# 회귀 모델 성능 평가

`lm`: 회귀 모델 적합 결과로 평가 지표를 확인할 수 있다

```r
library(car)
```

Multiple R-squared가 0.6534 이므로 회귀모형의 설명력은 65.3%이다. 즉, speed가 distance를 65.3% 설명하고 있다. 

```r
cars_index <- caret::createDataPartition(cars$speed, p = 0.8)
cars_train <- cars[unlist(cars_index), ]
cars_test <- cars[-unlist(cars_index), ]

cars_train_lm <- lm(dist ~ speed, data = cars_train)
summary(cars_train_lm)
```

```r
cars_predict <- predict(cars_train_lm, newdata = cars_test)
cars_predict
```

MSE는 126.8732 정도이다.

```r
mean((cars_test$dist - cars_predict)**2)
```
