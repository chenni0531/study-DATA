```python
# 필요한 라이브러리 불러오기

import pandas as pd
import numpy as np
import random
import warnings
warnings.filterwarnings("ignore")

from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('../input/sp500-daily-19862018/spx.csv')
data['date'] = pd.to_datetime(data['date'])
data.set_index('date', inplace=True)
data.head()
```

```python
data['close'].plot(kind='line', figsize=(20,6))
```

## 1. White Noise

```python
# wn = np.random.normal(data['close'].mean(), data['close'].std(), size = len(data['close']))
wn = np.random.normal(0, data['close'].std(), size = len(data['close']))
data['wn'] = wn


plt.figure(figsize=(20, 6))
plt.plot(data['wn'])
plt.title("White Noise", fontsize=25)
plt.show()
```

## 2. Random Walk

1부터 시작해서 -1 또는 1을 랜덤하게 선택하고 전의 값에 더하는 방식으로 랜덤 워크를 구현한다.

```python
random_walk = []
random_walk.append(-1 if random.random() <0.5 else 1)

for i in range(1,1000):
    direction = -1 if random.random() <0.5 else 1
    value = random_walk[i - 1] + direction
    random_walk.append(value)

plt.figure(figsize=(20, 6))
plt.plot(random_walk)
plt.title("Random walk", fontsize=25)
plt.show()
```
